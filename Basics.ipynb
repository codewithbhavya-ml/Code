6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'


```python
marks=int(input("Enter the marks"))
if marks>=90:
    print("A")
elif marks>=75 and marks<89:
    print("B")
elif marks>=50 and marks<74:
    print("C")
else:
    print("Fail")

```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
Student=pd.DataFrame({'Name':['Raju','Pk','Rakesh','Rajesh','Ramya','Rachita','Uv','Manu','Anu','Pk'],
                      'Age':[25,26,24,35,21,19,22,20,23,26],
                      'City':['Mangalore','Udupi','Mumbai','Rajastan','Andra','Up','Kerala','Mp','Goa','Udupi'],
                      'Salary':[20000,19000,56000,np.nan,10000,11000,np.nan,34000,15000,19000]})
print(Student)
```

```python
Student.info() #information.
```

```python
Student.describe() #descriptive.
```

```python
Student_1=Student.drop('City',axis=1) #dropping the column City.
print(Student_1)
```

```python
import warnings
warnings.filterwarnings('ignore') #removing warning messages.

Student['Salary'].fillna(Student['Salary'].mean(),inplace=True) #filling null values.
print(Student)
```

```python
Student=Student.drop_duplicates() #removing duplicate values.
print(Student)
```
