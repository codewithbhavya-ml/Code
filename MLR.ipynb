```python
#Model building 
Y=data['Price']
X=data[['Weight']]
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
model=LinearRegression()
model.fit(X,Y)
data['Y_pred']=model.predict(X)
r2=r2_score(Y,data['Y_pred'])
mse=mean_squared_error(Y,data['Y_pred'])
rmse=np.sqrt(mse)
print("RMSE:",np.round(rmse,2))
print("R2 Score:",np.round(r2,2))

```

```python
import statsmodels.formula.api as smf
model=smf.ols('Price ~ Weight',data=data).fit()

r2=model.rsquared
print("R squared:",r2)

rmse=np.sqrt(np.mean(model.resid**2))
print("RMSE:",rmse)  

model.summary()
```

1.What is Normalization & Standardization and how is it helpful?

Normalization rescales the values of features to a fixed range, typically [0, 1].
Standardization transforms the data so it has a mean = 0 and standard deviation = 1.

1.Prevents features with large values from dominating others.

2.Speeds up convergence in optimization algorithms (like Gradient Descent).

2.What techniques can be used to address multicollinearity in multiple linear regression?

1.Remove highly correlated variables using correlation matrix or VIF.

2.Combine correlated features into a single variable.

3.Use PCA to convert correlated variables into uncorrelated components.

4.Apply regularization (Ridge or Lasso regression).

5.Center variables when using polynomial or interaction terms.
